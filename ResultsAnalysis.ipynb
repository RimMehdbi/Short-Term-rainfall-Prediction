```python
def Plot_subPlot(df, axs, year, display_legend= False, metric = "RMSE (mm)"):
  # Proposed= df.iloc[:, 4]
  LSTM = df.iloc[:,3]
  ARIMA = df.iloc[:,2]
  Persistence= df.iloc[:,1]
  xlabels = df.iloc[:,0]
  rects1 = axs.bar(ind, Persistence, width, color='b')
  rects2 = axs.bar(ind + width, ARIMA, width, color='orange')
  rects3 = axs.bar(ind + 2*width, LSTM, width, color='g')
  #rects4 = axs.bar(ind + 3*width, Proposed , width, color='r')
  axs.set_xticks(ind + width)
  axs.set_xticklabels(xlabels,fontsize=14)
  axs.set_xlabel("Lead time", fontsize=14)
  axs.set_ylabel(metric, fontsize=14)
  axs.set_title('Models performance in terms of '+ metric)
  if (display_legend):
     axs.legend((rects1[0], rects2[0],rects3[0]),('Persistence', 'ARIMA','LSTM'),bbox_to_anchor = (0, 1.01), fontsize= 12)
  labelvalue(rects1 , axs)
  labelvalue(rects2, axs)
  labelvalue(rects3, axs)
  axs.set_title(year)   
  #(0.175, 1.01)
```
